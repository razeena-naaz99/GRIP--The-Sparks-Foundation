### Visualization

```python
# Distribution of Sales,Discount,Profit,Quantity
sns.set(rc={"figure.figsize": (10,10)})
subplot(2,2,1)
sns.distplot(df['Sales'])
plt.title("Distribution of Sales")
subplot(2,2,2)
sns.distplot(df['Discount'])
plt.title("Distribution of Discount")
subplot(2,2,3)
sns.distplot(df['Profit'])
plt.title("Distribution of Profit")
subplot(2,2,4)
sns.distplot(df['Quantity'])
plt.title("Distribution of Quantity")
```

```python
sns.set(rc={"figure.figsize": (15,28)})
subplot(4,2,1)
sns.countplot(df["Ship Mode"])
plt.title("Ship Mode")
subplot(4,2,2)
labels= list(df['State'].unique())
chart = sns.countplot(df["State"])
chart.set_xticklabels(labels=labels, rotation=90)
plt.title("State")
subplot(4,2,3)
sns.countplot(df['Segment'])
plt.title("Segment")
subplot(4,2,4)
sns.countplot(df['Category'])
plt.title("Category")
subplot(4,2,5)
labels= list(df['Sub-Category'].unique())
x=sns.countplot(df['Sub-Category'])
x.set_xticklabels(labels=labels, rotation=90)
plt.title("Sub-Category")
subplot(4,2,6)
sns.countplot(df['Region'])
plt.title("Region")
subplot(4,2,7)
sns.countplot(df['Discount'])
plt.title("Discount")
# labels = list(df['Region'].unique())
# w=sns.countplot(df['Region'])
# w.set_xticklabels(labels=labels, rotation=45)
# plt.title("Outlet Type Count")
```

```python
plt.style.use('fivethirtyeight')
sns.set(rc={"figure.figsize": (8,8)})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
fig,axes=plt.subplots(7,1,figsize=(15,40))
sns.scatterplot(x='Category',y='Sales',hue='Profit',ax=axes[0],data=df)
sns.scatterplot(x='Sub-Category',y='Sales',hue='Profit',ax=axes[1],data=df)
labels= list(df['State'].unique())
x=sns.scatterplot(x='State',y='Sales',hue='Profit',ax=axes[2],data=df)
x.set_xticklabels(labels=labels, rotation=90)
sns.scatterplot(x='Segment',y='Sales',hue='Profit',ax=axes[3],data=df)
sns.scatterplot(x='Ship Mode',y='Sales',hue='Profit',ax=axes[4],data=df)
sns.scatterplot(x='Region',y='Sales',hue='Profit',ax=axes[5],data=df)
sns.scatterplot(x='Discount',y='Sales',hue='Profit',ax=axes[6],data=df)
```

```python
df.hist(figsize=(15,12),alpha = 0.3,bins = 50)
```

```python
df1=df.groupby('State').agg({'Sales':np.mean})
df_1=pd.DataFrame(df1["Sales"].sort_values())
fig=plt.figure(figsize=(10,15))
plt.barh(df_1.index,df_1['Sales'], color = 'orange')
```

```python
df1=df.groupby('State').agg({'Profit':np.mean})
df_1=pd.DataFrame(df1["Profit"].sort_values())
fig=plt.figure(figsize=(10,15))
plt.barh(df_1.index,df_1['Profit'], color = 'orange')
```
